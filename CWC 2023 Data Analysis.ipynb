22. Calculate the overall Team Performance of Each Team

```python
def team_performance(India):
    team_matches = data[(data["Team A"] == India) | (data["Team B"] == team_name)]
    
    # Total Matches Played
    total_matches = len(team_matches)
    
    # Wins and Win Percentage
    wins = len(team_matches[team_matches["Wining Team"] == team_name])
    win_percentage = (wins / total_matches) * 100 if total_matches > 0 else 0
    
    # Total Runs Scored (as Team A or Team B)
    total_runs = team_matches.loc[team_matches["Team A"] == India, "Score A"].sum() + \
                 team_matches.loc[team_matches["Team B"] == India, "Score B"].sum()
    
    avg_runs_per_match = total_runs / total_matches if total_matches > 0 else 0

    # Total Wickets Taken
    total_wickets_taken = team_matches.loc[team_matches["Team A"] == India, "Wickets B"].sum() + \
                          team_matches.loc[team_matches["Team B"] == India, "Wickets A"].sum()
    
    # Display performance summary
    performance_summary = {
        "Team": India,
        "Matches Played": total_matches,
        "Matches Won": wins,
        "Win Percentage": round(win_percentage, 2),
        "Total Runs Scored": total_runs,
        "Average Runs per Match": round(avg_runs_per_match, 2),
        "Total Wickets Taken": total_wickets_taken
    }

    return performance_summary
```

## Win/Loss Ratio of each team

```python
TeamWin = data["Wining Team"].value_counts()
TeamLoss = data.apply(lambda row: row["Team A"] if row["Wining Team"] == row["Team B"] else row["Team B"], axis=1).value_counts()
Win_Loss_ratio = (TeamWin / TeamLoss).reset_index()
Win_Loss_ratio.columns = ["Team", "Win/Loss Ratio"]

fig = p.bar(
    Win_Loss_ratio,
    x="Team",
    y="Win/Loss Ratio",
    title="Win/Loss Ratio of Teams",
    labels={"Win/Loss Ratio": "Win/Loss Ratio", "Team": "Teams"},
    color="Win/Loss Ratio",
    color_continuous_scale="Blues", height=400, width=1000,
)

fig.update_layout(xaxis_title="Team", yaxis_title="Win/Loss Ratio", template="plotly_dark")
fig.write_html("Flask Application/Visualisation/Win-Loss Ratio of Each Team.html")
fig.show()
```
