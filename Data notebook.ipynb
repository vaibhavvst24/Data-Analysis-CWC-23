7.What is the win/loss ratio for each team?

```python
TeamWin = data['Wining Team'].value_counts()
TeamLoss = data.apply(lambda row: row['Team A'] if row['Wining Team'] == row['Team B'] else row['Team B'], axis=1).value_counts()
Win_Loss_ratio = TeamWin / TeamLoss
print(Win_Loss_ratio)
```

```python
fig = plt.figure(figsize=(13,6), dpi=120, facecolor = 'k', edgecolor='cyan')
plt.title('Win/Loss Ratio', color='w')
plt.plot(TeamWin, color='firebrick', marker='h', linestyle="--", linewidth=2, markersize=12)
plt.plot(TeamLoss, color='darkslateblue',marker='p', linestyle="--", linewidth=2, markersize=12)
plt.xlabel('TeamWin', size=20, color='w')
plt.ylabel('TeamLoss', size=20, color='w')
plt.tick_params(color='w', labelcolor='w')
plt.savefig('images/Win-Loss Ratio.png', dpi=300, bbox_inches='tight')
fig = px.bar(data, x='Man of the Match', y='Wining Team', title="Win\Loss Ratio")
fig.write_html('visulization/Win-Loss Ratio.html')
plt.grid()
plt.show()
```
